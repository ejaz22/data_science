```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y,'*',label='sin')
plt.plot(x,z,'--',label='cos')

plt.title('sin cos curve')
plt.legend()
```

```python
plt.scatter(x,y)
```
